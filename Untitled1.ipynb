```python
import pandas as pd
```

```python
train_data = pd.read_csv('train.csv')
train_data
```

```python
x = train_data.iloc[:,2:8]
x
```

```python
y = train_data.iloc[:,8]
y
```

```python
from sklearn.ensemble import RandomForestRegressor
Rfr = RandomForestRegressor()
Rfr.fit(x,y)
```

```python
test_data = pd.read_csv('test.csv')
test_data
```

```python
x_p = test_data.iloc[:,2:8]
x_p
```

```python
pred = pd.DataFrame(Rfr.predict(x_p))
pred
```

```python
test_data['id'] = test_data['id'].to_csv('submission.csv')
```

```python
sub = pd.read_csv('submission.csv')
sub
```

```python
sub['num_orders'] = pred
```

```python
sub = sub.to_csv('final_sub.csv')
```
